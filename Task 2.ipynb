1) Color Quantization Function (# of Clusters K as input)

```python
def quantization(img,k): 
    w, h, d = original_shape = tuple(img.shape)
    image_array = np.reshape(img, (w*h, d))
    image_array_sample = shuffle(image_array, random_state=0)[:1000]
    kmeans = KMeans(n_clusters=k, random_state=0).fit(image_array_sample)
    labels = kmeans.predict(image_array)
    codebook_random = shuffle(image_array, random_state=0)[:k + 1]
    labels_random = pairwise_distances_argmin(codebook_random, image_array,axis = 0)
    codebook = kmeans.cluster_centers_
    d = codebook.shape[1]
    image = np.zeros((w,h,d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            image[i][j] = codebook[labels[label_idx]]
            label_idx += 1
    plt.imshow(image)

    pass

quantization(img,100)
```
